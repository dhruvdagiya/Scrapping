```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
from time import sleep
from random import randint
```

```python
headers = {"Accept-Language": "en-US,en;q=0.5"}
```

```python
#Declaring the variables

poster_link = []
movie_name = []
rating = []
genre = []
description = []
director = []
stars = []
```

```python
pages = np.arange(1,6820, 50)
```

```python
for page in pages:
  page = requests.get("https://www.imdb.com/search/title/?title_type=feature&num_votes=10000,&countries=us&sort=user_rating,desc&start="+str(page)+"&ref_=adv_nxt")
  soup = BeautifulSoup(page.text, 'html.parser')
  movies_data = soup.findAll('div', attrs = {'class': 'lister-item mode-advanced'})
  sleep(randint(2,8))

  for store in movies_data:
    poster = store.img['loadlate']
    poster_link.append(poster)

    name = store.h3.a.text
    movie_name.append(name)

    rate = store.strong.text
    rating.append(rate)

    gen = store.p.find('span', class_ = 'genre').text.replace(',',' ').replace('\n', '').strip()
    genre.append(gen)

    describe = store.find_all('p', class_ = 'text-muted')
    description_ = describe[1].text.replace('\n', '') if len(describe) >1 else '*****'
    description.append(description_)

    value_d = store.findAll("p")[2:]
    value_d_ = value_d[0].text.split('|')[0].replace('\n','').replace('Director:','').replace('Directors:','').strip()
    director.append(value_d_)

    value_s = store.findAll("p")[2:]
    value_s_ = value_s[0].text.split('|')[0].replace('\n','').replace('Director:','').replace('Directors:','').strip()
    stars.append(value_s_)
```

```python
IMDB_Movies = pd.DataFrame({'Poster': poster_link,'Name_Movie':movie_name, 'Rating': rating, 'Genre': genre, 'Description': description, 'Director': director, 'Cast': stars})
```

```python
IMDB_Movies.to_csv('IMDB_Movies.csv')
```
